````python
"""
## 1. æ ‘å½¢é€‰æ‹©é—®é¢˜ - DFSå›æº¯

### é—®é¢˜æè¿°
ç»™å®šä¸€æ£µæ ‘ï¼Œæ¯ä¸ªèŠ‚ç‚¹æœ‰ä¸€ä¸ªä»·å€¼ã€‚è¦æ±‚é€‰æ‹©kä¸ªèŠ‚ç‚¹ï¼Œä½¿å¾—è¿™kä¸ªèŠ‚ç‚¹åŠå…¶æ‰€æœ‰ç›¸é‚»èŠ‚ç‚¹çš„ä»·å€¼ä¹‹å’Œæœ€å¤§ã€‚

### è§£é¢˜æ€è·¯
- **DFSæœç´¢**ï¼šéå†æ‰€æœ‰å¯èƒ½çš„kä¸ªèŠ‚ç‚¹ç»„åˆ
- **å›æº¯**ï¼šå°è¯•é€‰æ‹©ä¸€ä¸ªèŠ‚ç‚¹åï¼Œé€’å½’å¤„ç†å‰©ä½™èŠ‚ç‚¹
- **çŠ¶æ€ç®¡ç†**ï¼šç»´æŠ¤å·²é€‰æ‹©èŠ‚ç‚¹ã€å·²è®¿é—®èŠ‚ç‚¹ã€å½“å‰ä»·å€¼

### å…³é”®éš¾ç‚¹
1. é€‰æ‹©ä¸€ä¸ªèŠ‚ç‚¹åï¼Œå…¶æ‰€æœ‰ç›¸å…³èŠ‚ç‚¹ä¹Ÿè¢«"å ç”¨"
2. éœ€è¦åœ¨DFSä¸­æ­£ç¡®ç®¡ç†çŠ¶æ€å’Œå›æº¯
3. æœ€ç»ˆæ¯”è¾ƒæ‰€æœ‰å¯èƒ½çš„kä¸ªèŠ‚ç‚¹ç»„åˆä¸­çš„æœ€å¤§å€¼

### å¤æ‚åº¦
- æ—¶é—´ï¼šO(C(n,k) * k) - ç»„åˆæ•° Ã— å¤„ç†æ—¶é—´
- ç©ºé—´ï¼šO(n) - é€’å½’æ ˆæ·±åº¦

### åº”ç”¨åœºæ™¯
- å›¾çš„ç‹¬ç«‹é›†é—®é¢˜
- èµ„æºé€‰æ‹©ä¼˜åŒ–é—®é¢˜
"""

import sys

class TreeNode:
    def __init__(self, val):
        self.val = val
        self.related = [self]  # ç›¸å…³èŠ‚ç‚¹åˆ—è¡¨

class Solution:
    def treeProblem(self, nodes, k):
        """
        é€‰æ‹©kä¸ªèŠ‚ç‚¹ä½¿å¾—ç›¸å…³èŠ‚ç‚¹æ€»ä»·å€¼æœ€å¤§
        
        Args:
            nodes: æ‰€æœ‰TreeNodeå¯¹è±¡åˆ—è¡¨
            k: è¦é€‰æ‹©çš„èŠ‚ç‚¹æ•°é‡
            
        Returns:
            int: æœ€å¤§ä»·å€¼å’Œ
        """
        self.res = -1
        self.path = 0
        self.visited = []
        self.chosen = []
        
        def dfs():
            if len(self.chosen) == k:
                self.res = max(self.res, self.path)
                return
            
            options = [a for a in nodes if a not in self.visited]
            for option in options:
                visited_copy = self.visited.copy()
                chosen_copy = self.chosen.copy()
                
                self.chosen.append(option)
                for r in option.related:
                    self.visited.append(r)
                    
                self.path += option.val
                dfs()
                
                self.path -= option.val
                self.visited = visited_copy
                self.chosen = chosen_copy
        
        dfs()
        return self.res

# ç¤ºä¾‹
# nodes = [TreeNode(5), TreeNode(3), TreeNode(7)]
# solution = Solution()
# result = solution.treeProblem(nodes, 2)
# print(f"æœ€å¤§ä»·å€¼: {result}")
````

````python
"""
## 2. å­—ç¬¦é‡å¤è®¡æ•° - åŠ¨æ€è§„åˆ’åŸºç¡€

### é—®é¢˜æè¿°
ç»™å®šä¸€ä¸ªå­—ç¬¦ä¸²ï¼Œè®¡ç®—æ¯ä¸ªä½ç½®å¤„è¯¥å­—ç¬¦åœ¨ä¹‹å‰æ‰€æœ‰å‡ºç°çš„æ¬¡æ•°ã€‚

### è§£é¢˜æ€è·¯
ä½¿ç”¨DPæ•°ç»„è®°å½•ç´¯ç§¯å€¼ï¼š
- `res[i]` = åˆ°ç¬¬iä½ç½®çš„ç´¯ç§¯é‡å¤æ¬¡æ•°
- `res[i] = res[i-1] + (i+1) - count(s[0:i+1], s[i])`
  - `(i+1)`: åˆ°ç¬¬iä½ç½®å…±æœ‰i+1ä¸ªå­—ç¬¦
  - `count()`: è¯¥å­—ç¬¦åœ¨å‰i+1ä¸ªå­—ç¬¦ä¸­å‡ºç°çš„æ¬¡æ•°
  - `res[i-1]`: å‰ä¸€ä¸ªä½ç½®çš„ç´¯ç§¯å€¼

### ç¤ºä¾‹åˆ†æ
```
s = "abcabc"
i=0: 'a' ç¬¬1æ¬¡ï¼Œcount=1ï¼Œres[0] = 0
i=1: 'b' ç¬¬1æ¬¡ï¼Œcount=1ï¼Œres[1] = 0 + 2 - 1 = 1
i=2: 'c' ç¬¬1æ¬¡ï¼Œcount=1ï¼Œres[2] = 1 + 3 - 1 = 3
i=3: 'a' ç¬¬2æ¬¡ï¼Œcount=2ï¼Œres[3] = 3 + 4 - 2 = 5
```

### å…³é”®ç‚¹
- è¿™æ˜¯DPçš„ç®€å•åº”ç”¨
- æ¯æ­¥åªä¾èµ–å‰ä¸€æ­¥çŠ¶æ€
- éœ€è¦å¿«é€Ÿè®¡ç®—å­—ç¬¦å‡ºç°æ¬¡æ•°ï¼ˆå¯ç”¨Counterä¼˜åŒ–ï¼‰

### å¤æ‚åº¦
- æ—¶é—´ï¼šO(nÂ²) - è®¡ç®—countæ—¶æ‰«æå‰ç¼€
- ç©ºé—´ï¼šO(n) - DPæ•°ç»„
"""

import sys

n = int(sys.stdin.readline().strip()) if hasattr(sys.stdin, 'readline') else 6
s = str(sys.stdin.readline().strip()) if hasattr(sys.stdin, 'readline') else "abcabc"

res = [0] * n
for i in range(1, n):
    c = s[i]  # å½“å‰å­—ç¬¦
    # è®¡ç®—è¯¥å­—ç¬¦åœ¨å‰i+1ä¸ªå­—ç¬¦ä¸­çš„å‡ºç°æ¬¡æ•°
    count = s[:i+1].count(c)
    res[i] = res[i-1] + (i+1) - count

# è¾“å‡º
for i in range(n):
    print(f"ä½ç½®{i}: å­—ç¬¦='{s[i]}', ç´¯ç§¯é‡å¤={res[i]}")
````

````python
"""
## 4. ç¡¬å¸å…‘æ¢ (Coin Change) - è®°å¿†åŒ–æœç´¢

### é—®é¢˜æè¿°
ç»™å®šä¸åŒé¢å€¼çš„ç¡¬å¸å’Œç›®æ ‡é‡‘é¢ï¼Œæ±‚æœ€å°‘éœ€è¦å¤šå°‘æšç¡¬å¸ã€‚

### è§£é¢˜æ–¹æ³•ï¼šè®°å¿†åŒ–é€’å½’ (Top-Down DP)

### é€’æ¨å…³ç³»
coinChange(amount) = 1 + min(coinChange(amount - coin) for coin in coins)

### åŸºå‡†æƒ…å†µ
- amount == 0: è¿”å›0ï¼ˆæ— éœ€ç¡¬å¸ï¼‰
- amount < 0: è¿”å›-1ï¼ˆä¸å¯è¡Œï¼‰

### ä¼˜åŒ–
- ä½¿ç”¨memoæ•°ç»„ç¼“å­˜å·²è®¡ç®—ç»“æœï¼Œé¿å…é‡å¤è®¡ç®—
- åˆå§‹åŒ–memoä¸ºç‰¹æ®Šå€¼-666è¡¨ç¤ºæœªè®¡ç®—

### å¤æ‚åº¦
- æ—¶é—´å¤æ‚åº¦: O(amount * len(coins))
- ç©ºé—´å¤æ‚åº¦: O(amount)
"""

# æ³¨æ„ï¼šéœ€è¦å…ˆå®šä¹‰amountå’Œcoins
amount = 5  # ç¤ºä¾‹ï¼šç›®æ ‡é‡‘é¢
coins = [1, 2, 5]  # ç¤ºä¾‹ï¼šç¡¬å¸é¢å€¼
memo = [-666] * (amount + 1)

def coinChange(coins, amount):
    """
    è®¡ç®—å…‘æ¢ç»™å®šé‡‘é¢æ‰€éœ€çš„æœ€å°‘ç¡¬å¸æ•°
    
    Args:
        coins: ç¡¬å¸é¢å€¼åˆ—è¡¨
        amount: ç›®æ ‡é‡‘é¢
        
    Returns:
        int: æœ€å°‘ç¡¬å¸æ•°ï¼Œå¦‚æœæ— æ³•å…‘æ¢è¿”å›-1
    """
    if amount == 0:
        return 0
    if amount < 0:
        return -1
    if memo[amount] != -666:
        return memo[amount]  # è¿”å›ç¼“å­˜ç»“æœ
    
    res = float('inf')
    for coin in coins:
        subProblem = coinChange(coins, amount - coin)
        if subProblem == -1:
            continue
        res = min(res, subProblem + 1)  # +1è¡¨ç¤ºä½¿ç”¨ä¸€æšè¯¥ç¡¬å¸
    
    # ç¼“å­˜ç»“æœ
    memo[amount] = res if res != float('inf') else -1
    return memo[amount]

# æµ‹è¯•
# result = coinChange(coins, amount)
# print(f"éœ€è¦ {result} æšç¡¬å¸æ¥å…‘æ¢ {amount}")
````

````python
"""
## 4. çŸ©é˜µå•è¯æœç´¢ - DFSå›æº¯

### é—®é¢˜æè¿°
åœ¨çŸ©é˜µä¸­æœç´¢æ˜¯å¦å­˜åœ¨æŒ‡å®šçš„å•è¯è·¯å¾„ã€‚å•è¯å¿…é¡»ç”±ç›¸é‚»å•å…ƒæ ¼ï¼ˆä¸Šä¸‹å·¦å³ï¼‰ä¸­çš„å­—æ¯è¿æ¥ç»„æˆã€‚

### è§£é¢˜æ€è·¯
1. éå†çŸ©é˜µæ‰¾åˆ°ç›®æ ‡å­—ç¬¦çš„èµ·ç‚¹
2. ä»èµ·ç‚¹å¼€å§‹DFSæœç´¢ç›®æ ‡å•è¯çš„å®Œæ•´è·¯å¾„
3. æ ‡è®°å·²è®¿é—®å•å…ƒæ ¼é˜²æ­¢é‡å¤ä½¿ç”¨
4. é€’å½’æœç´¢å››ä¸ªæ–¹å‘

### å…³é”®æŠ€å·§
- **åŸåœ°æ ‡è®°**ï¼šç”¨Noneæ ‡è®°è®¿é—®è¿‡çš„å•å…ƒæ ¼ï¼Œè€Œéé¢å¤–æ•°ç»„
- **åŠæ—¶æ¢å¤**ï¼šé€’å½’è¿”å›å‰æ¢å¤å•å…ƒæ ¼åŸå€¼
- **è¾¹ç•Œæ£€æŸ¥**ï¼šé˜²æ­¢è¶Šç•Œå’Œéæ³•è®¿é—®

### åº”ç”¨åœºæ™¯
- å•è¯æœç´¢æ¸¸æˆ
- æ–‡æœ¬åŒ¹é…é—®é¢˜
- å›¾ä¸Šè·¯å¾„æœç´¢

### å¤æ‚åº¦
- æ—¶é—´ï¼šO(m*n*4^L) - mÃ—nçŸ©é˜µï¼Œæœ€å¤š4^Lç§è·¯å¾„ï¼ˆL=å•è¯é•¿åº¦ï¼‰
- ç©ºé—´ï¼šO(L) - é€’å½’æ·±åº¦
"""

def find_target(matrix, target):
    """åœ¨çŸ©é˜µä¸­æŸ¥æ‰¾ç›®æ ‡å•è¯"""
    if not matrix or not target:
        return False
    
    rows, cols = len(matrix), len(matrix[0])
    
    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] == target[0]:
                if search(matrix, target, i, j, 0):
                    return True
    return False

def search(matrix, target, i, j, index):
    """
    DFSæœç´¢å‡½æ•°
    
    Args:
        matrix: å­—ç¬¦çŸ©é˜µ
        target: ç›®æ ‡å•è¯
        i, j: å½“å‰ä½ç½®åæ ‡
        index: ç›®æ ‡å•è¯çš„å½“å‰ç´¢å¼•
    """
    # åŸºå‡†æƒ…å†µï¼šå·²åŒ¹é…å®Œæ•´å•è¯
    if index == len(target):
        return True
    
    # è¾¹ç•Œå’Œå­—ç¬¦æ£€æŸ¥
    if (i < 0 or i >= len(matrix) or 
        j < 0 or j >= len(matrix[0]) or 
        matrix[i][j] != target[index]):
        return False
    
    # æ ‡è®°ä¸ºå·²è®¿é—®
    temp = matrix[i][j]
    matrix[i][j] = None
    
    # å››æ–¹å‘æœç´¢
    result = (search(matrix, target, i - 1, j, index + 1) or  # ä¸Š
              search(matrix, target, i + 1, j, index + 1) or  # ä¸‹
              search(matrix, target, i, j - 1, index + 1) or  # å·¦
              search(matrix, target, i, j + 1, index + 1))    # å³
    
    # æ¢å¤
    matrix[i][j] = temp
    return result

# æµ‹è¯•
# matrix = [['a', 'b', 'c'], ['e', 'e', 'f'], ['f', 'h', 'i']]
# print(find_target(matrix, 'eefihe'))
````

````python
"""
## 3. æ’åˆ—é—®é¢˜ - å›æº¯æ³•

### é—®é¢˜æè¿°
ç”Ÿæˆä¸€ä¸ªæ•°ç»„çš„æ‰€æœ‰æ’åˆ—ç»„åˆã€‚

### è§£é¢˜æ€è·¯ï¼šå›æº¯æ³• (Backtracking)
1. ç»´æŠ¤å½“å‰è·¯å¾„ (track) å’Œå·²ä½¿ç”¨æ ‡è®° (used)
2. é€’å½’æ„å»ºæ’åˆ—
3. å½“è·¯å¾„é•¿åº¦ç­‰äºæ•°ç»„é•¿åº¦æ—¶ï¼Œæ‰¾åˆ°ä¸€ä¸ªå®Œæ•´æ’åˆ—
4. æ¯æ¬¡å°è¯•åéœ€è¦æ¢å¤çŠ¶æ€ (æ’¤é”€é€‰æ‹©)

### å…³é”®æ¦‚å¿µ
- **åšé€‰æ‹© (Choose)**: å°è¯•ä¸€ä¸ªæ•°å­—åŠ å…¥æ’åˆ—
- **è¿›å…¥ä¸‹å±‚ (Explore)**: é€’å½’æ„å»ºå‰©ä½™éƒ¨åˆ†
- **æ’¤é”€é€‰æ‹© (Unchoose)**: æ¢å¤çŠ¶æ€ç»§ç»­å°è¯•

### æ¨¡æ¿åº”ç”¨
æ­¤æ¨¡æ¿å¯æ‰©å±•åˆ°ï¼š
- ç»„åˆé—®é¢˜ (né€‰k)
- å­é›†ç”Ÿæˆ
- æ£‹ç›˜é—®é¢˜ç­‰

### å¤æ‚åº¦
- æ—¶é—´ï¼šO(n! * n) - n!ç§æ’åˆ— Ã— æ¯ç§æ‹·è´næ¬¡
- ç©ºé—´ï¼šO(n) - é€’å½’æ·±åº¦
"""

from typing import List

class Solution:
    def __init__(self):
        self.res = []
       
    def permute(self, nums: List[int]) -> List[List[int]]:
        """
        ç”Ÿæˆæ•°ç»„çš„æ‰€æœ‰æ’åˆ—
        
        Args:
            nums: ä¸é‡å¤çš„æ•°å­—æ•°ç»„
            
        Returns:
            List[List[int]]: æ‰€æœ‰æ’åˆ—
        """
        track = []
        used = [False] * len(nums)
        self.backtrack(nums, track, used)
        return self.res

    def backtrack(self, nums: List[int], track: List[int], used: List[bool]) -> None:
        """å›æº¯é€’å½’å‡½æ•°"""
        if len(track) == len(nums):
            self.res.append(track.copy())
            return
            
        for i in range(len(nums)):
            if used[i]:
                continue
            
            # åšé€‰æ‹©
            track.append(nums[i])
            used[i] = True
            
            # é€’å½’
            self.backtrack(nums, track, used)
            
            # æ’¤é”€é€‰æ‹©
            track.pop()
            used[i] = False

# æµ‹è¯•
# solution = Solution()
# print(solution.permute([1, 2, 3]))
````

````python
"""
## 3.5 æˆæœ¬æœ€å°åŒ–é—®é¢˜ - å›æº¯ + å‰ªæ

### é—®é¢˜æè¿°
æœ‰nä¸ªè®¾å¤‡ï¼Œæ¯ä¸ªè®¾å¤‡æœ‰æ”¶ç›Š(profit)å’Œæˆæœ¬(cost)ã€‚
è¦æ±‚ï¼šé€‰æ‹©è‹¥å¹²è®¾å¤‡ä½¿å¾—æ€»æ”¶ç›Šâ‰¥1ï¼ŒåŒæ—¶æœ€å°åŒ–æ€»æˆæœ¬ã€‚

### è§£é¢˜æ€è·¯
- **å›æº¯æ³•**ï¼šå°è¯•é€‰æ‹©/ä¸é€‰æ‹©æ¯ä¸ªè®¾å¤‡
- **å‰ªæä¼˜åŒ–**ï¼šå¦‚æœå½“å‰æˆæœ¬å·²è¶…è¿‡æœ€å°å€¼ï¼Œåˆ™åœæ­¢æœç´¢è¯¥åˆ†æ”¯
- **è´ªå¿ƒç­–ç•¥**ï¼šé€šè¿‡ç»´æŠ¤æœ€å°æˆæœ¬æ¥å‰ªæ

### å…³é”®æŠ€å·§
1. **æ—©åœ**ï¼šä¸€æ—¦æ”¶ç›Šâ‰¥1ç«‹å³è¿”å›
2. **å‰ªææ¡ä»¶**ï¼š`if pay >= self.minPay: return`
3. **å›æº¯çŠ¶æ€**ï¼šæ­£ç¡®ç®¡ç†visitedåˆ—è¡¨

### å¤æ‚åº¦
- æœ€åï¼šO(2^n) - ä¸å‰ªæ
- å®é™…ï¼šO(2^k) - kä¸ºæœ‰æ•ˆåˆ†æ”¯æ•°ï¼ˆå‰ªææ•ˆæœå–å†³äºæ•°æ®ï¼‰

### ä¸åŠ¨æ€è§„åˆ’å¯¹æ¯”
- DPï¼šæ›´é€‚åˆæ±‚æœ€ä¼˜å€¼çš„æ€»æ•°
- å›æº¯ï¼šæ›´é€‚åˆæ±‚æœ€ä¼˜è§£æœ¬èº«
"""

import sys

class Solution:
    def solveProblem(self):
        """æ±‚æœ€å°æˆæœ¬"""
        n = int(sys.stdin.readline())
        cost = [float('inf')] * n
        profit = [0] * n
        for i in range(n):
            a, b = map(int, sys.stdin.readline().split())
            profit[i] = a
            cost[i] = b
        
        self.minPay = 9999
        visited = []
        
        def backtrack(pay, elec):
            """
            Args:
                pay: å½“å‰æ€»æˆæœ¬
                elec: å½“å‰æ€»æ”¶ç›Š
            """
            # æå‰ç»ˆæ­¢ï¼šæ”¶ç›Šè¶³å¤Ÿ
            if elec >= 1:
                self.minPay = min(self.minPay, pay)
                return
            
            # å‰ªæï¼šå½“å‰æˆæœ¬å·²è¶…è¿‡å·²çŸ¥æœ€å°å€¼
            if pay >= self.minPay:
                return
            
            options = [i for i in range(n) if i not in visited]
            for option in options:
                pay += cost[option]
                elec += profit[option]
                visited.append(option)
                
                backtrack(pay, elec)
                
                pay -= cost[option]
                elec -= profit[option]
                visited.pop()
        
        backtrack(0, 0)
        print(1 if self.minPay == 9999 else self.minPay)

# æµ‹è¯•ï¼ˆéœ€è¦æ ‡å‡†è¾“å…¥ï¼‰
# Solution().solveProblem()
````

````python
"""
## 9. èºæ—‹çŸ©é˜µéå† - æ•°ç»„æ¨¡æ‹Ÿ

### é—®é¢˜æè¿°
æŒ‰èºæ—‹é¡ºåºéå†mÃ—nçŸ©é˜µå¹¶è¿”å›æ‰€æœ‰å…ƒç´ ã€‚èºæ—‹é¡ºåºï¼šä»å¤–å‘å†…ï¼Œå³â†’ä¸‹â†’å·¦â†’ä¸Šã€‚

### éå†è¿‡ç¨‹
1. ä»å·¦åˆ°å³éå†é¡¶è¡Œ
2. ä»ä¸Šåˆ°ä¸‹éå†å³åˆ—
3. ä»å³åˆ°å·¦éå†åº•è¡Œï¼ˆéœ€æ£€æŸ¥æ˜¯å¦è¿˜æœ‰è¡Œï¼‰
4. ä»ä¸‹åˆ°ä¸Šéå†å·¦åˆ—ï¼ˆéœ€æ£€æŸ¥æ˜¯å¦è¿˜æœ‰åˆ—ï¼‰
5. é‡å¤ç›´åˆ°éå†å®Œæ‰€æœ‰å…ƒç´ 

### å…³é”®éš¾ç‚¹
- **è¾¹ç•Œç®¡ç†**ï¼šç»´æŠ¤å››ä¸ªè¾¹ç•Œï¼ˆtop, bottom, left, rightï¼‰
- **æ¡ä»¶æ£€æŸ¥**ï¼šé¿å…é‡å¤éå†æŸä¸€è¡Œæˆ–åˆ—
- **è¿­ä»£é¡ºåº**ï¼šç¡®ä¿æ¯æ¬¡è¿­ä»£éƒ½æ­£ç¡®æ”¶ç¼©è¾¹ç•Œ

### ç›¸ä¼¼é¢˜ç›®
- è›‡å½¢éå†
- åˆ†å±‚éå†ï¼ˆå¥‡å¶å±‚ï¼‰
- å¯¹è§’çº¿éå†

### å¤æ‚åº¦
- æ—¶é—´ï¼šO(m*n) - è®¿é—®æ¯ä¸ªå…ƒç´ ä¸€æ¬¡
- ç©ºé—´ï¼šO(m*n) - ç»“æœæ•°ç»„
"""

def spiralOrder(matrix):
    """
    æŒ‰èºæ—‹é¡ºåºéå†çŸ©é˜µ
    
    Args:
        matrix: mÃ—nçŸ©é˜µ
        
    Returns:
        list: èºæ—‹éå†çš„å…ƒç´ åˆ—è¡¨
    """
    if not matrix or not matrix[0]:
        return []
    
    rows, cols = len(matrix), len(matrix[0])
    result = []
    
    top, bottom = 0, rows - 1
    left, right = 0, cols - 1
    
    while top <= bottom and left <= right:
        # 1. ä»å·¦åˆ°å³éå†é¡¶è¡Œ
        for i in range(left, right + 1):
            result.append(matrix[top][i])
        top += 1
        
        # 2. ä»ä¸Šåˆ°ä¸‹éå†å³åˆ—
        for i in range(top, bottom + 1):
            result.append(matrix[i][right])
        right -= 1
        
        # 3. ä»å³åˆ°å·¦éå†åº•è¡Œï¼ˆæ£€æŸ¥è¾¹ç•Œï¼‰
        if top <= bottom:
            for i in range(right, left - 1, -1):
                result.append(matrix[bottom][i])
            bottom -= 1
        
        # 4. ä»ä¸‹åˆ°ä¸Šéå†å·¦åˆ—ï¼ˆæ£€æŸ¥è¾¹ç•Œï¼‰
        if left <= right:
            for i in range(bottom, top - 1, -1):
                result.append(matrix[i][left])
            left += 1
    
    return result

# æµ‹è¯•
# matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
# print(spiralOrder(matrix))  # [1, 2, 3, 6, 9, 8, 7, 4, 5]
````

````python
"""
## 8. è´¨å› æ•°åˆ†è§£ - ç»„åˆæ•°å­¦ + å¿«é€Ÿå¹‚

### é—®é¢˜æè¿°
ç»™å®šæ•´æ•°nå’Œyï¼Œæ±‚æ»¡è¶³ä»¥ä¸‹æ¡ä»¶çš„xçš„æ–¹æ¡ˆæ•°ï¼š
- xæ˜¯yçš„å› å­
- 1 â‰¤ x â‰¤ y
- ç»“æœæ¨¡10^9+7

### è§£é¢˜æ€è·¯

#### æ­¥éª¤1ï¼šè´¨å› æ•°åˆ†è§£
å°†yåˆ†è§£ä¸ºç´ æ•°çš„ä¹˜ç§¯ï¼šy = pâ‚^kâ‚ * pâ‚‚^kâ‚‚ * ... * pâ‚˜^kâ‚˜

#### æ­¥éª¤2ï¼šè®¡ç®—ç»„åˆæ•°
å¯¹yçš„æ¯ä¸ªè´¨å› å­påŠå…¶æŒ‡æ•°kï¼š
- å¯¹äºxä¹Ÿæ˜¯yçš„å› å­ï¼Œpåœ¨xä¸­çš„æŒ‡æ•°å¯ä»¥æ˜¯ 0, 1, 2, ..., k
- å…±æœ‰ k+1 ç§é€‰æ‹©
- å¯¹äºnä¸ªç‹¬ç«‹é€‰æ‹©ï¼Œæ–¹æ¡ˆæ•° = (k+1)^n

#### æ­¥éª¤3ï¼šä¹˜æ³•åŸç†
æ€»æ–¹æ¡ˆæ•° = âˆ((káµ¢+1)^n) å¯¹æ‰€æœ‰i

### å…³é”®ä¼˜åŒ–
- **å¿«é€Ÿå¹‚**ï¼šè®¡ç®—(káµ¢+1)^n mod (10^9+7)
- **è´¨å› æ•°åˆ†è§£**ï¼šO(âˆšy)éå†æ‰¾æ‰€æœ‰è´¨å› å­

### å¤æ‚åº¦
- æ—¶é—´ï¼šO(âˆšy + m*logn) - è´¨å› æ•°åˆ†è§£ + å¿«é€Ÿå¹‚
- ç©ºé—´ï¼šO(m) - è´¨å› æ•°ä¸ªæ•°

### åº”ç”¨åœºæ™¯
- å› å­è®¡æ•°é—®é¢˜
- ç»„åˆæ•°å­¦
- æ•°è®ºé—®é¢˜
"""

import sys

MOD = 10**9 + 7

def power(base, exp):
    """å¿«é€Ÿå¹‚ï¼šè®¡ç®—(base^exp) % MOD"""
    res = 1
    base %= MOD
    while exp > 0:
        if exp % 2 == 1:
            res = (res * base) % MOD
        base = (base * base) % MOD
        exp //= 2
    return res

def get_prime_factorization(num):
    """
    è´¨å› æ•°åˆ†è§£
    
    Returns:
        dict: {è´¨å› å­: æŒ‡æ•°}
    """
    factors = {}
    d = 2
    temp = num
    while d * d <= temp:
        while temp % d == 0:
            factors[d] = factors.get(d, 0) + 1
            temp //= d
        d += 1
    if temp > 1:
        factors[temp] = factors.get(temp, 0) + 1
    return factors

def solve():
    """å¤„ç†å•ä¸ªæµ‹è¯•ç”¨ä¾‹"""
    n, y = map(int, sys.stdin.readline().split())
    
    # è´¨å› æ•°åˆ†è§£
    prime_factors = get_prime_factorization(y)
    
    total_ways = 1
    # å¯¹yçš„æ¯ä¸€ä¸ªè´¨å› æ•°åŠå…¶æŒ‡æ•°
    for p, k in prime_factors.items():
        # è®¡ç®— (k+1)^n
        term1 = power(k + 1, n)
        # è®¡ç®— k^n
        term2 = power(k, n)
        
        # è¿™ä¸ªè´¨å› å­å¯¹åº”çš„æ–¹æ¡ˆæ•°ï¼š(k+1)^n - k^n
        ways_for_p = (term1 - term2 + MOD) % MOD
        
        # ä¹˜ä»¥æ€»æ–¹æ¡ˆæ•°
        total_ways = (total_ways * ways_for_p) % MOD
    
    print(total_ways)

# ä¸»å‡½æ•°
def main():
    """å¤„ç†å¤šç»„æµ‹è¯•æ•°æ®"""
    t = int(sys.stdin.readline())
    for _ in range(t):
        solve()

if __name__ == "__main__":
    main()

# æµ‹è¯•ç¤ºä¾‹
# è¾“å…¥ï¼šn=2, y=12  (12 = 2^2 * 3^1)
# æ–¹æ¡ˆæ•° = (2+1)^2 * (1+1)^2 - ... = 9 * 4 = 36
````
